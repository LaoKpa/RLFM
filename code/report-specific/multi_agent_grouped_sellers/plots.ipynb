```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
csvs = [
    "multi_seller_group_iter500_run0.csv",
    "multi_seller_group_iter500_run1.csv",
    "multi_seller_group_iter500_run2.csv",
    "multi_seller_group_iter500_run3.csv"
]
```

```python
dfs = [pd.read_csv(csv, index_col='Episodes') for csv in csvs]
```

```python
fig = plt.figure()
for i, df in enumerate(dfs):
    ax = plt.subplot(220 + i +1)
    legend = (i==1)
    df[['B1', 'B2', 'B3', 'Sellers']].plot(ax=ax, legend=legend)
    if i+1 in [1,3]:
        plt.ylabel("Mean rewards")

fig.set_size_inches(w=4.7747, h=3.5)
plt.savefig('grouped_sellers.pgf')
```
