```python
fig = plt.figure()
for i, df in enumerate(dfs):
    ax = plt.subplot(220 + i +1)
    legend = (i+1==2)
    df[['B1', 'B2', 'B3', 'S1', 'S2', 'S3']].plot(ax=ax, legend=legend)
    if i+1 in [1,3]:
        plt.ylabel("Mean rewards")

fig.set_size_inches(w=4.7747, h=4.)
plt.savefig('multi_3v3_info.pgf')
```
